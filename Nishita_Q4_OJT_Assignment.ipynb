Q4. Polynomial Plotting with Annotations:


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set random seed
np.random.seed(37)

# Randomly select the coefficients of a 5th degree polynomial
coefficients = np.random.randint(-10, 11, size=6)

# Define the polynomial function
def polynomial(x):
    return np.polyval(coefficients, x)

# Generate x values
x_values = np.linspace(-5, 5, 400)

# Generate y values using the polynomial function
y_values = polynomial(x_values)

# Plot the polynomial
plt.plot(x_values, y_values, label='Polynomial')

# Find maxima and minima
maxima_indices = np.r_[True, y_values[1:] > y_values[:-1]] & np.r_[y_values[:-1] > y_values[1:], True]
minima_indices = np.r_[True, y_values[1:] < y_values[:-1]] & np.r_[y_values[:-1] < y_values[1:], True]

# Annotate maxima
for i in range(len(x_values[maxima_indices])):
    plt.annotate('Max', xy=(x_values[maxima_indices][i], y_values[maxima_indices][i]), xytext=(-20, 20),
                 textcoords='offset points', arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

# Annotate minima
for i in range(len(x_values[minima_indices])):
    plt.annotate('Min', xy=(x_values[minima_indices][i], y_values[minima_indices][i]), xytext=(-20, -20),
                 textcoords='offset points', arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('5th Degree Polynomial with Maxima and Minima')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
